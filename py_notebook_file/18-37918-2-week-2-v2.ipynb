simple calculator using if statements

```python
x= input("Enter Value 1st: ")
y= input("Enter Value 2nd: ")

print("Chose an operation: ")
print("1--> +")
print("2--> -")
print("3--> *")
print("4--> /")
sgn= input()

if sgn=='1':
        result=int(x)+int(y)
        print("Result:", result)
       

elif sgn=='2':
        result=int(x)-int(y)
        print("Result:", result)
            

elif sgn=='3':
        result=int(x)*int(y)
        print("Result:", result)
        

elif sgn=='4':
        result=int(x)/int(y)
        print("Result:", result)
        
        

    
```
